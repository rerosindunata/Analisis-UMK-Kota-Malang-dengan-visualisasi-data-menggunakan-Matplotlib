# Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Data Loading

Data UMK di Kota Malang bersumber dari BPS Kota Malang dan Disnakertrans Prov Jawa Timur.

```python
data = pd.read_excel('um_malkot.xlsx')
```

```python
print(data)
```

# Data Visualisasi

```python
plt.figure(figsize=(15,5))

plt.plot(data['Tahun'], data['Rupiah'], label='Rupiah', color = 'blue', marker = 'o', linestyle='dashed')

plt.xlabel('Tahun')
plt.ylabel('Rupiah (dalam Juta)')

plt. title('UMK di Kota Malang Tahun 2000 - 2022 (dalam Rupiah)', size = 15)

plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.plot(data['Tahun'], data['Persen'], label='Persen', color = 'red', marker = 'o', linestyle='dashed')

plt.xlabel('Tahun')
plt.ylabel('Persen')

plt. title('UMK di Kota Malang tahun 2000 - 2022 (dalam Persen)', size = 15)

plt.show()
```

Data UMK di Kota Malang tahun 2000 - 2022.
1. Setiap tahun UMK di Kota Malang mengalami kenaikan dengan nilai yang beragam. Kecuali pada tahun 2002 tidak terjadi kenaikan, nominalnya sama dengan 2001 yaitu Rp325.000,-.
2. Kenaikan tertinggi terjadi pada tahun 2001 yaitu sebesar 37.7% lebih tinggi dari tahun 2000.
3. Kenaikan terendah terjadi pada tahun 2022 sebesar 0.79% dan pada tahun 2021 sebesar 2.59%. Hal ini mungkin terjadi karena tahun 2021 dan 2022 masih adanya pandemi Covid-19 yang mengakibatkan ekonomi menjadi terdampak.
4. Setiap kali terjadi masalah ekonomi maka kenaikan nilai UMK di Kota Malang akan menurun dari tahun - tahun sebelumnya. Selain karena Pandemi Covid-19 yang dimulai pada tahun 2020, dapat dilihat bahwa pada tahun 2008 terjadi penurunan yang signifikan dari pada tahun - tahun sebelumnya yang naik diatas 10%, pada saat terjadi krisis ekonomi 2008 kenaikan UMK di Kota Malang hanya 4.41%.
5. Setelah ekonomi mulai membaik maka kenaikan UMK di Kota Malang menjadi stabil kembali, bisa dilihat dari tahun 2013 hingga tahun 2020 kenaikannya stabil diatas 8% dan kembali menurun pada saat Pandemi Covid-19.
